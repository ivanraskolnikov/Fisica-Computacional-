```python
# Derivada de Euler ; derivada simetrica
```

```python
def de_E(f,x):   #derivada de Euler
    df=np.zeros_like(f)  #genera ceros
    for i in range (len(x)-1):    #len, longitud de 
        df[i] = (f[i+1]-f[i])/(x[i+1]-x[i])
    df[-1]=- (f[-2]-f[-1])/ (x[-1]-x[-2])  #sacar la derivada del √∫ltimo punto, revisar notas 
    return df

def de_S(f,x):   #derivada de Simetrica
    df=np.zeros_like(f)  #genera ceros
    for i in range (1,len(x)-1):    #len, longitud de 
        df[i] = (f[i+1]-f[i-1])/(x[i+1]-x[i-1])
    df[0]=- (f[1]-f[0])/ (x[1]-x[0])
    df[-1]=- (f[-2]-f[-1])/ (x[-1]-x[-2])  #sacar la derivada del √∫ltimo punto, revisar notas 
    return df
```

```python
x = np.linspace(0,2*np.pi)
y = np.sin(x)
dy_a = np.cos(x)
dy_n = de_E(y,x)
dy_S = de_S(y,x)
plt.plot(x,y,label='f(x)')
plt.plot(x,dy_a,label=r"$f'(x)$ analitica")
plt.plot(x,dy_n,label=r"$f'(x)$ numerica Euler")
plt.plot(x,dy_S, '-',label=r"$f'(x)$ numerica Simetrica")
plt.legend()
plt.show()

```
