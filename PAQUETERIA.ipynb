```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = [3,5,"iván"]
print (a[0],a[1]) #con esto imprimo solo algunos elementos 
print (a[-1])     #asi se invierte el sentido 
print (a)
```

```python
x = np.linspace(0,2*np.pi,100)  #(minimo, máximo, numero de particion)
y = np.tan(x)     
y1= np.sin(x)
y2= np.cos(x)
y[:-1][np.diff(y)<0]=np.nan   #-1 no toma el ultimo, se hace para que se pueda tener un array con cardinalidad práctica
plt.plot (x,y, label=r'$tan(x)$') #grafica 
plt.plot (x,y1, label=r'$sin(x)$')
plt.plot (x,y2, label=r'$cos(x)$')
plt.ylim(-2, 2)   #cambio de limites en el eje y
plt.title (r'Grafica de funciones trigonometricas') #la r es para que reconozca el texto como codigo latex
plt.xlabel (r'$x$'), plt.ylabel (r'$y$')
plt.savefig('grafica tan.pdf')
plt.show()
```
