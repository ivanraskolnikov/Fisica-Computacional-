```python
import numpy as np
import folium as fl
import matplotlib.pyplot as plt
from IPython.display import display, IFrame
import os #ésta librería se comunica con el sistema op. puede funcionar como una terminal, con ciertas restricciones 
```

```python
def mi_lista_csv(varios_archivos):   #ésta función filtra los archivos, para solo quedarnos con aquellos que tengan 
    filtro=[]                         #extensión csv
    for nombre in varios_archivos:
        if nombre[-4:]=='.csv':
            filtro.append(nombre)
    return filtro 
```

```python
def mi_mapa(archivo_mapa):    #función para crear un mapa de algun archivo
    datos_mapa = np.genfromtxt(archivo_mapa,delimiter=',',dtype=float)  #extracción de archivos
    t, lat_long = datos_mapa[1:,0], datos_mapa[1:,1:3]  #asignación de variables
    mapa = fl.Map(location=lat_long[0],zoom_start=16) #creación del mapa en el inicio
    trayectoria = fl.PolyLine(locations=lat_long).add_to(mapa)
    fl.Marker(location=[lat_long[0,0],lat_long[0,1]], 
              popup="Inicio", 
              icon=fl.Icon(color='green')).add_to(mapa)
    fl.Marker(location=[lat_long[-1,0],lat_long[-1,1]], 
              popup="Final", 
              icon=fl.Icon(color='red')).add_to(mapa)
    mapa.save('mapa.html')
```

```python
lista = os.listdir('./ArchivosCSV/') #arroja la lista de los nombres de la carpeta 
lista = mi_lista_csv(lista)
mi_mapa('./ArchivosCSV/'+ lista[0])

```

```python
def mi_mapa(archivo_mapa, ruta='./'):    #función para crear un mapa de algun archivo modificado
    datos_mapa = np.genfromtxt(ruta + archivo_mapa,delimiter=',',dtype=float)  #extracción de archivos
    t, lat_long = datos_mapa[1:,0], datos_mapa[1:,1:3]  #asignación de variables
    mapa = fl.Map(location=lat_long[0],zoom_start=16) #creación del mapa en el inicio
    trayectoria = fl.PolyLine(locations=lat_long).add_to(mapa)
    fl.Marker(location=[lat_long[0,0],lat_long[0,1]], 
              popup="Inicio", 
              icon=fl.Icon(color='green')).add_to(mapa)
    fl.Marker(location=[lat_long[-1,0],lat_long[-1,1]], 
              popup="Final", 
              icon=fl.Icon(color='red')).add_to(mapa)
    mapa.save(ruta + archivo_mapa[:-4]+'.html')
```

```python
ruta_csv =  './ArchivosCSV/'
lista = os.listdir(ruta_csv) #arroja la lista de los nombres de la carpeta 
lista = mi_lista_csv(lista)
for mapita in lista: 
    mi_mapa(mapita,ruta=ruta_csv)
    mapa_html = ruta_csv + mapita[:-4] + '.html'
    print(mapa_html)
    marco = IFrame(src=mapa_html, width=250, height=250)
    display(marco)
    
```
