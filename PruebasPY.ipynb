```python
x1= np.linspace(-5,5,200)
x_1= np.floor(x1)
y0=[]
for i in x_1:
    if i % 2 == 0:
        j=1
        y0.append(j+1)
    else:
        j=i
        y0.append(j)
y0=np.array(y0)


y0=np.array(y0)
y1=[]

for i in y0:
    if i % 2 != 0:
        j=1
        y1.append(j)
    else:
        j=i
        y1.append(j)
y1=np.array(y1)
y1=y1-1

y1[:-1][np.diff(y1)<0]=np.nan
y1[:-1][np.diff(y1)>0]=np.nan


plt.plot(x1,y1)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 
import scipy.integrate as integrate

def f(x,t):
        return (1/np.pi)*(np.cos((t**3)/3)+x*t)

x = np.linspace(-20,10,20)
y = []
for i in x:
    res1, _ = integrate.quad(lambda t: f(i,t), 0, 20)
    res2, _ = integrate.quad(lambda t: f(i,t), 20, 200)
    result = res1 + res2
    y.append(result)
y = np.array(y) 
```

```python
import numpy as np 
import matplotlib.pyplot as plt
n = np.array([3,5,10,20,30])
x = np.linspace(-np.pi,np.pi,200)

def S_n (x,n):
    S=0
    for i in range(0,n+1):
        S += (((-1)**i)/((2*i+1)**2))*np.sin((2*i+1)*x)
    return(S)

for i in n:
    y = S_n(x,i) 
    plt.plot(x,y,label=r'$n$ = %2.1f' %i)
    plt.legend()
    plt.grid()
    plt.show()
```

```python
from scipy.special import jv

n = np.array([0,1,2,3])
x = np.linspace(0,20,200)
y = np.zeros_like(x)

for i in n:
    j = jv(i,x)
    plt.plot(x,j,label=r'$n$ = %2.2f' %i)
plt.plot(x,y, color='black')
plt.legend()
plt.grid()
plt.show()
```

```python
from scipy.special import spherical_jn

n = np.array([0,1,2,3])
x = np.linspace(0,20,200)
y = np.zeros_like(x)

for i in n:
    j = spherical_jn(i,x)
    plt.plot(x,j,label=r'$n$ = %2.2f' %i)
plt.plot(x,y, color='black')
plt.legend()
plt.grid()
plt.show()
```
