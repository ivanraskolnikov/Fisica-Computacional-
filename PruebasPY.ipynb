```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(-4*np.pi,4*np.pi,100)
y = 1/np.tan(x)
y[:-1][np.diff(y)>0]=np.nan
plt.plot(x,y)
plt.ylim(-2, 2)  
plt.show()
```

```python
x1= np.linspace(-5,5,200)
x_1= np.floor(x1)
y0=[]
for i in x_1:
    if i % 2 == 0:
        j=1
        y0.append(j+1)
    else:
        j=i
        y0.append(j)
y0=np.array(y0)


y0=np.array(y0)
y1=[]

for i in y0:
    if i % 2 != 0:
        j=1
        y1.append(j)
    else:
        j=i
        y1.append(j)
y1=np.array(y1)
y1=y1-1

y1[:-1][np.diff(y1)<0]=np.nan
y1[:-1][np.diff(y1)>0]=np.nan


plt.plot(x1,y1)
plt.grid()
plt.show()
```
